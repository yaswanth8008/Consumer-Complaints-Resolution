```python
def report(results, n_top):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

If we need to submit the classes we need to choose a threshold value based on which we decide the class of the predicted value 

```python
cutoffs=np.linspace(0.01,0.99,99)

cutoffs
```

```python
train_score=logr.predict_proba(X_Train)[:,1]

real=Y_Train
```

```python
train_score
```

Let's select the threshold value of probability.Here the problem we are trying to solve is to mininmize the False Negatives,i.e to not miss the consumers who are likely to dispute.So the relavant evaluation metric would be recall.
we will use F-Beta score with higher value of beta to focus more on recall

```python
F_beta=[]

for cutoff in cutoffs:
    
    predicted=(train_score>cutoff).astype(int)

    TP=((predicted==1) & (real==1)).sum()
    TN=((predicted==0) & (real==0)).sum()
    FP=((predicted==1) & (real==0)).sum()
    FN=((predicted==0) & (real==1)).sum()
    
    Precision = TP / (TP+FP)
    Recall = TP / (TP+FN)
      
    f_beta_score = (5*Precision*Recall) / ((4*Precision) + Recall) #beta=2
    
    
    F_beta.append(f_beta_score)

```

```python
list(zip(cutoffs,F_beta))
```

```python
mycutoff=cutoffs[F_beta==max(F_beta)][0]
mycutoff
```

```python
test_classes=(test_score>mycutoff).astype(int)
```
